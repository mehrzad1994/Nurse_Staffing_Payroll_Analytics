```python
#make a list of all the columns names so that when we want to work with a subset of data we would be able to choose from it
#delete excess columns
columns_list = ['PROVNUM', 'PROVNAME', 'CITY', 'STATE', 'COUNTY_NAME', 'COUNTY_FIPS', 'CY_Qtr','WorkDate',
                'MDScensus', 'Hrs_RNDON', 'Hrs_RNDON_emp', 'Hrs_RNDON_ctr', 'Hrs_RNadmin', 'Hrs_RNadmin_emp', 
                'Hrs_RNadmin_ctr', 'Hrs_RN', 'Hrs_RN_emp', 'Hrs_RN_ctr', 'Hrs_LPNadmin', 'Hrs_LPNadmin_emp', 
                'Hrs_LPNadmin_ctr', 'Hrs_LPN', 'Hrs_LPN_emp', 'Hrs_LPN_ctr', 'Hrs_CNA', 'Hrs_CNA_emp',
                'Hrs_CNA_ctr', 'Hrs_NAtrn', 'Hrs_NAtrn_emp', 'Hrs_NAtrn_ctr', 'Hrs_MedAide', 'Hrs_MedAide_emp',
                'Hrs_MedAide_ctr']

#delete CY_Qtr column because all these data is belong to the last quarter
columns_list = ['PROVNUM', 'PROVNAME', 'CITY', 'STATE', 'COUNTY_NAME', 'COUNTY_FIPS', 'WorkDate',
                'MDScensus', 'Hrs_RNDON', 'Hrs_RNDON_emp', 'Hrs_RNDON_ctr', 'Hrs_RNadmin', 'Hrs_RNadmin_emp', 
                'Hrs_RNadmin_ctr', 'Hrs_RN', 'Hrs_RN_emp', 'Hrs_RN_ctr', 'Hrs_LPNadmin', 'Hrs_LPNadmin_emp', 
                'Hrs_LPNadmin_ctr', 'Hrs_LPN', 'Hrs_LPN_emp', 'Hrs_LPN_ctr', 'Hrs_CNA', 'Hrs_CNA_emp',
                'Hrs_CNA_ctr', 'Hrs_NAtrn', 'Hrs_NAtrn_emp', 'Hrs_NAtrn_ctr', 'Hrs_MedAide', 'Hrs_MedAide_emp',
                'Hrs_MedAide_ctr']

#cmsdata = cmsdata.drop(['CY_Qtr'],axis=1)

#create a new data frame based on the reduced columns list
cmsdata=cmsdata[columns_list]
#print number of rows and columns of the reduced data frame
print(cmsdata.shape)
```

```python
# Check number of categorical and numerical features

#make a list of categorical features and print it
categorical_features = cmsdata.select_dtypes(exclude = [np.number]).columns
print("Categorical features:\n",categorical_features)

#make a list of numerical features and print it
numerical_features = cmsdata.select_dtypes(include = [np.number]).columns
print("\n\n\nNumerical features:\n",numerical_features)
```

```python
# Review the datatypes

print("Data Format of each feature:\n \n",cmsdata.dtypes)
```

```python
# add 3 more column for total hour of all contarctors, all employees and sum of both

#create a list of all total hours
columns_list_total = ['Hrs_RNDON', 'Hrs_RNadmin', 'Hrs_RN', 'Hrs_LPNadmin', 'Hrs_LPN', 'Hrs_CNA', 'Hrs_NAtrn', 'Hrs_MedAide']

#create a list of all employees hours
columns_list_emp = ['Hrs_RNDON_emp', 'Hrs_RNadmin_emp', 'Hrs_RN_emp', 'Hrs_LPNadmin_emp', 'Hrs_LPN_emp', 'Hrs_CNA_emp',
                'Hrs_NAtrn_emp', 'Hrs_MedAide_emp']

#create a list of all contracts hours
columns_list_ctr = ['Hrs_RNDON_ctr', 'Hrs_RNadmin_ctr', 'Hrs_RN_ctr', 'Hrs_LPNadmin_ctr',  'Hrs_LPN_ctr', 'Hrs_CNA_ctr', 
                'Hrs_NAtrn_ctr', 'Hrs_MedAide_ctr']


# Get sum of each columns list for each row
cmsdata['Hrs_total'] = cmsdata[columns_list_total].sum(axis=1)
cmsdata['Hrs_total_emp'] = cmsdata[columns_list_emp].sum(axis=1)
cmsdata['Hrs_total_ctr'] = cmsdata[columns_list_ctr].sum(axis=1)


#print number of rows and columns of the new data frame with added 3 columns
print(cmsdata.shape)

#print first few rows of the data frame
print(cmsdata.head());

#print statistical features of every column
cmsdata.describe()
```

```python
#fill Nan values with 0
cmsdata = cmsdata.fillna(0)

#delete the rows with Nan values
#cmsdata = cmsdata.dropna()


#print number of rows and columns of the data frame after preprocessing
print(cmsdata.shape)
```

```python
#change the type of work date to datetime to work easier with it
cmsdata['WorkDate'] =  pd.to_datetime(cmsdata['WorkDate'], format='%Y%m%d')

# check the datatypes
print("Data Format of work date column:",cmsdata['WorkDate'].dtypes) 
#print the first few rows
cmsdata['WorkDate'].head()
```

```python
#############################1
# top providers in each state
# group by state, provider, sort by total hours

# create a new data frame based on needed columns for this recommendation

cmsdata1=cmsdata[['PROVNAME', 'STATE',
                  'Hrs_total', 'Hrs_total_emp', 'Hrs_total_ctr']]

#group the information by state, provider
#calculating the sum for each group
#sort the information based on total hours for each state 

cmsdata1=cmsdata1.groupby(['STATE','PROVNAME'], as_index = False).agg({'Hrs_total':'sum',
                        'Hrs_total_emp':'sum', 'Hrs_total_ctr':'sum'}).sort_values(['STATE','Hrs_total'] ,ascending=False)
cmsdata1.columns = ['STATE', 'PROVNAME','Hrs_total', 'Hrs_total_emp', 'Hrs_total_ctr']


#cmsdata1=cmsdata1.groupby(['STATE','PROVNAME']).sum(['Hrs_total'])
#cmsdata1=cmsdata1.sort_values(by=['STATE','Hrs_total'], ascending=False)


#show the first 20 rows
cmsdata1.head(20)

#print(cmsdata1.columns.values)

```

```python
#save the dataframe as a csv file
cmsdata1.to_csv('recommendation1.csv', index=False)
```

```python
#############################2

#calculating the maximum contract hours for each state 
#this table show the most profitable states
#these are the states that needs the most nursing staff
#these are the states that we can focus on to gain more profit

# create a new data frame based on needed columns for this recommendation

cmsdata2=cmsdata[['STATE', 'Hrs_total_ctr']]

#group the information by state
#calculating the max of total contractors for each group
#sort the information based on max total contract hours for each state 

cmsdata2=cmsdata2.groupby(['STATE'], as_index = False).agg({'Hrs_total_ctr':'max'}).sort_values(['Hrs_total_ctr'] ,ascending=False)
cmsdata2.columns = ['STATE','Hrs_total_ctr']


cmsdata2.head(50)
```

```python
#save the dataframe as a csv file
cmsdata2.to_csv('recommendation2.csv', index=False)
```

```python
#############################3
# dataframe only clipboard
# group by state, city
# compare contract, employee, total

cmsdata3=cmsdata[['PROVNAME', 'STATE', 
                  'Hrs_total', 'Hrs_total_emp', 'Hrs_total_ctr']]

#providers.to_csv('providers.csv', index=False) 

#cmsdata2.loc[cmsdata2['PROVNAME'] == 'clipboard']
#!!!!!!!!!!!!!!!!!!!!!!!! there is no data about clipboard in this dataset

#print(cmsdata3["PROVNAME"].value_counts())


#select all the rows where provider is PARKVIEW CARE CENTER
cmsdata3=cmsdata3.loc[cmsdata3['PROVNAME'] == 'PARKVIEW CARE CENTER']

#cmsdata3.head()

#group the information by state
#calculating the sum of total hours, total employee hours and total contract hours for each group
#sort the information based on total hours for each state 

cmsdata3=cmsdata3.groupby(['STATE'], as_index = False).agg({'Hrs_total':'sum',
                        'Hrs_total_emp':'sum', 'Hrs_total_ctr':'sum'}).sort_values(['Hrs_total'] ,ascending=False)
cmsdata3.columns = ['STATE','Hrs_total', 'Hrs_total_emp', 'Hrs_total_ctr']

cmsdata3.head(20)
```
